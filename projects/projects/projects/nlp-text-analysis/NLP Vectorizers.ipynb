NLP Fundamentals: Text Vectorization

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
corpus =[
    'thr cat sat on the mat',
    'the dog sat on the log',
    'cats and dogs are great'
]    
```

CountVectorizer(Bag of Words)

Concept: it creates a vocabulary of all unique words and simply counts how many times each word appears in the document. It focuses on frequency.

```python
#initialize the Vectorizer
count_vectorizer = CountVectorizer()
#Fit and Trasnform the corpus
count_matrix = count_vectorizer.fit_transform(corpus)
```

```python
words = count_vectorizer.get_feature_names_out()
df_count = pd.DataFrame(count_matrix.toarray(), columns = words)
```

```python
print("--- CountVectorizer Result ---")
print(df_count)
```

TfidVectorizer(Term Frequency-Inverse Document Frequency)

Concept: TF=IDF penalize words that appears too frequently across all documents(like 'thr', 'is', 'on',and reward words that are unique to specific documents.
TF(Term Frequency): How often a word appears in a ducment.  
IDF: How rare a word is across the entire corpus.

```python
#Initialize the Vectorizer
tfidf_vectorizer = TfidfVectorizer()
#fit and transform
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
```

```python
#Convert to DataFrame
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=words)
print("\n--- TfidfVectorizer Result ---")
print(df_tfidf.round(3))
```
